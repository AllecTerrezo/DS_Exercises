```python
# Convert to DataFrame
df = pd.DataFrame(list(person_heights.items()), columns=['Name', 'Height'])

# Define the target mean
target_mean = df['Height'].mean()
```

```python
# Function to find valid teams
def find_valid_teams(df, target_mean):
    np.random.seed(0)  # For reproducibility
    valid_teams = []

    # Loop until two valid teams are found
    while len(valid_teams) < 2:
        indices = np.random.choice(df.index, 5, replace=False)
        sample = df.loc[indices]
        if np.isclose(sample['Height'].mean(), target_mean, atol=1e-3):
            valid_teams.append(indices.tolist())

    return valid_teams
```

```python
# Get valid teams
valid_team_indices = find_valid_teams(df, target_mean)
print("Indices of valid teams:", valid_team_indices)
```

```python
# Retrieve team members based on indices
teams = {}
for i, indices in enumerate(valid_team_indices, start=1):
    team = df.iloc[indices]
    teams[f"Team {i}"] = dict(zip(team['Name'], team['Height']))
```

```python
# Outputting the team members and their heights
for team_name, members in teams.items():
    print(team_name + ":")
    for name, height in members.items():
        print(f"{name}: {height}")
```
